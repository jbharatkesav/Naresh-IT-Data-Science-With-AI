# Legend

```python
x16 = np.arange(1,8)
fig, ax = plt.subplots()
ax.plot(x16, x16**3)
ax.plot(x16, x16/5)
ax.plot(x16, x16*4)
ax.legend(['Normal', 'Fast', 'Slow'])
```

```python
x16 = np.arange(1, 8)

fig, ax = plt.subplots()

ax.plot(x16, x16**3, label='Normal')
ax.plot(x16, x16/5, label='Fast')
ax.plot(x16, x16*4, label='Slow')

ax.legend();
```

# Adding color

```python
x16 = np.arange(1, 8)

fig, ax = plt.subplots()

ax.plot(x16, x16**3, color='Red')
ax.plot(x16, x16/5, color='Black')
ax.plot(x16, x16*4, color='Blue')
```

```python
x16 = np.arange(1, 8)

fig, ax = plt.subplots()

ax.plot(x16, x16**3, 'c' , label = 'N')
ax.plot(x16, x16/5, 'y', label = 'T')
ax.plot(x16, x16*4, 'm', label = 'U')
plt.legend()
```
